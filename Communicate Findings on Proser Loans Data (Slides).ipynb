## The Distrebution of  Borrower Annual Percentage Rate.

Borrower APR mean is quit equal to it's median which suggests that it's normaly distrebuted. which can be found in the following histgram.
he distrebution is roughly normaly distrebuted as stated above but there are some unsual peaks at the right side at `0.35` to `0.36` of the graph that might have an impact on the shape is normally and not skewed.
we also have small peaks arounc `0.1`, `0.33`.

```python
plt.figure(figsize=(12, 7))
borrower_apr_bins = np.arange(
    df['BorrowerAPR'].min(), 
    df['BorrowerAPR'].max(), 
    0.01
)
sb.histplot(data=df, x='BorrowerAPR', bins=borrower_apr_bins)
plt.title('Distrebution of Borrower Annual Percentage Rate');
plt.xlabel('Borrower Annual Percentage Rate')
plt.ylabel('Count');
```

## Distrebution of Loan Amount

```python
plt.figure(figsize=(12, 7))
loan_amountbins = np.arange(
    df['LoanOriginalAmount'].min(), 
    df['LoanOriginalAmount'].max(), 
    300.
)
sb.histplot(data=df, x='LoanOriginalAmount', bins=loan_amountbins)
plt.title('Distrebution of Loan Amount');
plt.xlabel('Loan Amount $');
plt.ylabel('Count');
```

## Amount of Loans in each Borrower State

```python
from plotly import graph_objects as go

## the code provied to create the graph for the jupyter notebook
## and i just linked a pre saved version directly and the code is here

borrower_state_loan_counts = (
    df['BorrowerState']
    .value_counts().reset_index()
    .rename(columns={'index': 'State', 'BorrowerState': 'Count'})
)
borrower_state_loan_counts.head()

graph_obj = go.Choropleth(
    locations=borrower_state_loan_counts['State'],
    z = borrower_state_loan_counts['Count'],
    locationmode = 'USA-states',colorscale = 'Blues',
    colorbar_title = "Loan Counts",
)

fig = go.Figure(data=graph_obj)
fig.update_layout(
    title_text = 'Loan Count in each U.S State',geo_scope='usa',
)
fig.show()
```

## Relation between Prosper Rating and Borrower Annual Percentage Rate.
BorrowerAPR variation for each category is significantly differ for each other as for risky borrowers the APR is about 0.30 and for safe borrowers is 0.18 which is very large when comparing between them.

```python
base_color = sb.color_palette()[0]
prosper_rating_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
```

```python
plt.figure(figsize=(12, 7))

sb.boxplot(data=df, x='ProsperRating (Alpha)', y='BorrowerAPR', 
           color=base_color, order=prosper_rating_order)
plt.title('Relation between Prosper Rating and Borrower Annual Percentage Rate');
plt.xlabel('Prosper Rating %')
plt.ylabel('Borrower Annual Percentage Rate');
```

## Relation between the Income Range and Prosper Rating.

the ProsperRating has an impact on the IncomeRange and the number of loan associated with each.
as in **IncomeRanges** in (**25K:50K**, **50K:75K**, **75K:100K**) has a Prosper Rating as **C**.
and in the **>100K** we start to see that the Borrowers with **A** and **AA** increasing.

```python
income_range_order = ['Not employed', '$0', '$1-24,999', '$25,000-49,999', 
                      '$50,000-74,999', '$75,000-99,999', '$100,000+'
]
```

```python
plt.figure(figsize=(12, 7))
sb.countplot(data=df, x='IncomeRange', hue='ProsperRating (Alpha)', 
             palette='magma_r', order=income_range_order, hue_order=prosper_rating_order)
plt.xticks(rotation=20);
plt.ylim((0, 6500))
plt.legend(ncol=3, loc='upper left', title='Prosper Rating');
plt.title('Proser Rating Disterbution by the Income Range');
plt.xlabel('Income Range $');
plt.ylabel('Count');
```

## Relation between Loan Amount and Borrower Annual Percentage Rating.
The data has a Negative Correlation as the **LoanOriginalAmount** increases the **BorrowerAPR** tends to decrease.<br>
most of the loans lies between (**0**, **15K**) and small portion lies between the (**25K**, **35K**)

```python
plt.figure(figsize=(12, 8))
sb.regplot(data=df, x='LoanOriginalAmount', y='BorrowerAPR', 
           scatter_kws={'alpha':.5}, line_kws={'alpha': .5, 'color': 'r'}, x_jitter=300)
plt.title('Loan Amount ($) by. Borrower Annual Percentage Rate');
plt.xlabel('Loan Amount $');
plt.ylabel('Borrwer Annual Percentage Rating.');
```

## Relation between the Loan Amount and Borrower Annual Percentage Rate and effect of Prosper Rating.

**ProsperRating** has a great effect on the relation between **LoanOriginalAmount** and **BorrowerAPR** as the overall relation is **Negative** but the relation differ from one **ProsperRating** to another as it:<br>
 - **Negative** for: (HR ,E ,D ,C ,B)<br>
 - **Positive** for : (AA, A)<br>
 
which means that the clients with (**A**, **AA**) rating when the Loan Amount increases the BorrowerAPR increases.

```python
g = sb.FacetGrid(data=df, col='ProsperRating (Alpha)', col_wrap=4, 
                 margin_titles=True, col_order=prosper_rating_order)
g.map(sb.regplot, 'LoanOriginalAmount', 'BorrowerAPR',
      scatter_kws={'alpha': .2, 'color': '#83add3'})
plt.text(x=.5, y=1.1, ha='center', fontsize=15, 
         s='Loan Amount vs. BorrowerAPR and effect of ProsperRating');
```

## Borrower Annual Percentage Rate vs. Proser Rating and Term

For **ProsperRating** relation with **BrowerAPR** we see intersting outcomes as
- the **Term** increase for Borrowers with (**A**, **AA**, **B**) we see increase in teh **BorrowerAPR**.
- and for the (**C**,**D**,**E**) we that is opposite of that as it started to decrease as the **Term** increased

```python
fig = plt.figure(figsize = [12,7])
sb.pointplot(data = df, x = 'ProsperRating (Alpha)', 
                  y = 'BorrowerAPR', hue = 'Term', palette = 'Reds', 
                  linestyles = '', dodge = .3, ci='sd', order=prosper_rating_order
)
plt.title('Borrower Annual Percentage Rate by  ProsperRating and Term')
plt.ylabel('Borrower Annual Percentage Rate');
plt.xlabel('Prosper Rating');
plt.legend(ncol=3, title='Term');
```
