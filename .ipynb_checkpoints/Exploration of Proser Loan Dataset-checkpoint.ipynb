### What features in the dataset do you think will help support your investigation into your feature(s) of interest?

I think than these variable below will have a great impact on the the two variable above.

 - **BorrowerAPR**: The Borrower's Annual Percentage Rate (APR) for the loan.
 - **Term**: The length of the loan expressed in months.
 - **LoanStatus**: The current status of the loan: Cancelled,  Chargedoff, Completed, Current, Defaulted, FinalPaymentInProgress, PastDue. The PastDue status will be accompanied by a delinquency bucket.
 - **BorrowerState**:The two letter abbreviation of the state of the address of the borrower at the time the Listing was created.
 - **EmploymentStatus**: The employment status of the borrower at the time they posted the listing.
 - **TotalTrade**: Number of trade lines ever opened at the time the credit profile was pulled.
 - **LoanOriginalAmount**: The origination amount of the loan.
 - **IncomeRange**: The income range of the borrower at the time the listing was created.
 - **StatedMonthlyIncome**: The monthly income the borrower stated at the time the listing was created.
 - **InquiriesLast6Months**: Number of inquiries in the past six months at the time the credit profile was pulled.
 - **DebtToIncomeRatio**: The debt to income ratio of the borrower at the time the credit profile was pulled. This value is Null if the debt to income ratio is not available. This value is capped at 10.01 (any debt to income ratio larger than 1000% will be returned as 1001%).
 - **EstimatedLoss**: Estimated loss is the estimated principal loss on charge-offs. Applicable for loans originated after July 2009.
 - **EstimatedReturn**: The estimated return assigned to the listing at the time it was created. Estimated return is the difference between the Estimated Effective Yield and the Estimated Loss Rate. Applicable for loans originated after July 2009.
 - **OnTimeProsperPayments**: Number of on time payments the borrower had made on Prosper loans at the time they created this listing. This value will be null if the borrower has no prior loans.
 - **ProsperGrade**: The Prosper Rating assigned at the time the listing was created, between AA - HR. The rating assigned by Prosper helps to decide the APR the borrower should be charged.
 - **ProsperRating (Alpha)**: The Prosper Rating assigned at the time the listing was created between AA - HR.  Applicable for loans originated after July 2009..<br>
 
all these info can be fount on the excel file included with the zip file or on this link [Here](https://www.prosper.com/Downloads/Services/Documentation/ProsperDataExport_Details.html)

```python
## select the variable (features) to work with in the analysis
features = [
    'BorrowerAPR', 'Term', 'BorrowerState', 'EmploymentStatus', 'LoanStatus','LoanOriginalAmount', 
    'IncomeRange', 'StatedMonthlyIncome', 'InquiriesLast6Months', 'DebtToIncomeRatio', 'EstimatedLoss', 
    'EstimatedReturn', 'OnTimeProsperPayments', 'ProsperRating (Alpha)', 'ListingKey',
]

df_features = df[features]
df_features
```

```python
df_features.info()
```

```python
df_features.isna().sum()
```

```python
## drop any duplicates rows from the selected dataframe
df_features = df_features.drop_duplicates()
df_features.shape
```

```python
## as my goal here is to work with data to figure out the ProsperRating (Alpha) columns
## I will drop any null values in this column from entire dataset.

df_features = df_features[~df_features['ProsperRating (Alpha)'].isna()]

## fill the NaN values in DebtIncomeRatio with the median as the distrebution
## will be right skewed as we will show down in the Univariate Visuals
df_features['DebtToIncomeRatio'].fillna(df_features['DebtToIncomeRatio'].median(), inplace=True)
df_features.head()
```

```python
df_features.info()
```

```python
df_features.isna().sum()
```
